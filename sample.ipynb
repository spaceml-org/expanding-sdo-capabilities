```python
import numpy as np
import matplotlib.pyplot as plt
from sdoml import sdo_read
```

```python
a = sdo_read(2010, 9, 2, 20, 0, subsample=4)
b = sdo_read(2010, 9, 2, 20, 0, instr='HMI', channel='by', subsample=4)
```

```python
plt.figure(figsize=(25,25))
plt.imshow(a,cmap='gray',clim=(0,20),origin='lower')
```

```python
plt.figure(figsize=(25,25))
plt.imshow(b,cmap='seismic_r',clim=(-3000,3000),origin='lower')
plt.imshow(a,alpha=0.2,cmap=plt.cm.RdBu_r)
```
