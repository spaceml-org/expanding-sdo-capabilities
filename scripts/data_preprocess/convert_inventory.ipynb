Script that will convert from the old IBM cluster paths in the inventory.pkl to updated new paths

```python
import pandas as pd

from sdo.global_vars import DATA_BASEDIR, DATA_INVENTORY
```

```python
df = pd.read_pickle(DATA_INVENTORY)
```

```python
df
```

```python
in_path = '/gpfs/gpfs_gl4_16mb/b9p111/fdl_sw/SDOMLmm/'
out_path = DATA_BASEDIR + '/'
```

```python
df.replace({in_path: out_path}, regex=True, inplace=True)
```

```python
df
```

```python
df.to_pickle(DATA_INVENTORY)
```
