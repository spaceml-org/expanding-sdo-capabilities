# Loading predictions and timestamps


In this notebook we want to provide an example of how to load the saved predictions and timestamp


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results_path = '/home/Valentina/EXPERIMENT_RESULTS/VIRTUAL_TELESCOPE/vale_test_gc/'
filename = results_path + '0002_vale_test_gc_train_predictions_timestamps.npz'
```

```python
outputs = np.load(filename)
```

```python
outputs.files
```

### Select based on index

```python
indexes = [0, 1, 2]
gt_outputs = outputs['gt_img'][indexes]
pr_outputs = outputs['img'][indexes]
timestamps = outputs['ts'][indexes]
```

```python
timestamps
```

```python
fig = plt.figure()
pos = 0
for i in range(len(indexes)):
    pos += 1
    ax = fig.add_subplot(2*len(indexes), 2, pos)      
    fig.subplots_adjust(top=3.0)
    ax.set_title(f'GT at time {timestamps[i][:4]}')
    ax.axis('off')
    ax.imshow(gt_outputs[i], vmin=0, vmax=1)
    pos += 1
    ax = fig.add_subplot(2*len(indexes), 2, pos)      
    fig.subplots_adjust(top=3.0)
    ax.set_title(f'Out at time {timestamps[i][:4]}')
    ax.axis('off')
    ax.imshow(pr_outputs[i], vmin=0, vmax=1)
```

## Select based on date

```python
def find_date_index(date, data):
    for index, item in enumerate(data['ts']):
        if all(item == date):
            return index
```

```python
date = [2012,    2,   10,   18,    0]
index = find_date_index(date, outputs)
print(index, outputs['ts'][index])
```
